# Урок 2. Анализ данных и проверка статистических гипотез.

1. В чём различие между зависимыми и независимыми выборками?
2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

1. Если можно установить, что каждому случаю из одной выборки соответствует один и только один случай из второй выборки и наоборот для каждого случая в двух выборках, такие выборки называются зависимыми. В случае, если такая взаимосвязь между выборками отсутствует, то эти выборки считаются независимыми.

Зависимые выборки имеют одинаковый объём, а объём независимых может отличаться.

2. Параметрический критерий используют при известных свойства выборки: тип распределения, среднее, дисперсия и т.д. В ином случае используют непараметрический критерий.
